```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

## Point slope form

```python
x = [-73.925,-73.94]
y = [40.79, 40.83]
```

```python
m = (y[1] - y[0]) / (x[1]-x[0])
```

```python
ys = np.arange(40.76,40.86,0.001)
```

```python
xs = (ys - y[0])/m + x[0]
```

$$ y - 40.79 = -8/3 (x + 73.925)$$ 

$$ y = -8/3x - 156.343 $$

```python
ground_df = pd.read_csv("Training_data_uhi_index_UHI2025-v2.csv")
ground_df.head()
```

```python
fig, ax = plt.subplots()
ax.plot(ground_df['Longitude'],ground_df['Latitude'],'.')
ax.plot(xs,ys)
plt.show()
```

```python
data = ground_df
```

```python
data['Area'] = np.where(
    data['Latitude'] > (-8/3) * data['Longitude'] - 156.343,
    'Bronx',
    'Manhattan'
)
```

```python
bronx = data.loc[data['Area'] == 'Bronx']
manhattan = data.loc[data['Area'] == 'Manhattan']
```

```python
fig2, ax2 = plt.subplots()
ax2.plot(bronx['Longitude'],bronx['Latitude'],'.')
ax2.plot(manhattan['Longitude'],manhattan['Latitude'],'.')
ax2.set_ylim(ax.get_ylim())
ax2.set_xlim(ax.get_xlim())
plt.show()
```

```python
manhattan = data.loc[data['Area'] == 'Manhattan'].copy()
manhattan_weather_data = pd.read_excel(io='NY_Mesonet_Weather.xlsx',sheet_name='Manhattan')

bronx = data.loc[data['Area'] == 'Bronx'].copy()
bronx_weather_data = pd.read_excel(io='NY_Mesonet_Weather.xlsx',sheet_name='Bronx')
```

```python
manhattan['datetime'] = pd.to_datetime(manhattan['datetime'], format='%d-%m-%Y %H:%M')
manhattan['datetime'] = manhattan['datetime'].dt.strftime('%Y-%m-%d %H:%M')
manhattan['datetime'] = pd.to_datetime(manhattan['datetime'])
manhattan = manhattan.sort_values(by='datetime')


bronx['datetime'] = pd.to_datetime(bronx['datetime'], format='%d-%m-%Y %H:%M')
bronx['datetime'] = bronx['datetime'].dt.strftime('%Y-%m-%d %H:%M')
bronx['datetime'] = pd.to_datetime(bronx['datetime'])
bronx = bronx.sort_values(by='datetime')
```

```python
bronx_weather_data.columns = ['datetime','airtemp','relative humidity', 'avg wind speed', 'wind direction', 'solar flux']
bronx_weather_data['datetime'] = bronx_weather_data['datetime'].str.replace('EDT','')
bronx_weather_data['datetime'] = pd.to_datetime(bronx_weather_data['datetime'])

manhattan_weather_data.columns = ['datetime','airtemp','relative humidity', 'avg wind speed', 'wind direction', 'solar flux']
manhattan_weather_data['datetime'] = manhattan_weather_data['datetime'].str.replace('EDT','')
manhattan_weather_data['datetime'] = pd.to_datetime(manhattan_weather_data['datetime'])
```

```python
bronx_final = pd.merge_asof(bronx, bronx_weather_data, on='datetime', direction='nearest')
man_final = pd.merge_asof(manhattan, manhattan_weather_data, on='datetime', direction='nearest')
```

```python
final_weather = pd.concat([bronx_final,man_final])
```
