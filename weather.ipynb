```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
ground_df = pd.read_csv("Training_data_uhi_index_UHI2025-v2.csv")
ground_df.head()
```

```python
x = [-73.925,-73.94]
y = [40.79, 40.83]
```

```python
m = (y[1] - y[0]) / (x[1]-x[0])
```

```python
y - y[0] = m (x - x[0])
```

```python
ys = np.arange(40.76,40.86,0.001)
```

```python
xs = (ys - y[0])/m + x[0]
```

```python
fig, ax = plt.subplots()
ax.plot(ground_df['Longitude'],ground_df['Latitude'],'.')
ax.plot(xs,ys)
```
